```python
def eda(df,nombre):
  print(f'EDA de {nombre}')
  print('Primeras filas')
  print(df.head())
  print()
  print(df.info())
  print()
  print('Filas repetidas',df.duplicated().sum())
  print()
  print('Valores ausentes')
  print(df.isna().sum())
  print()
```

```python
# EDA
eda(raw_partidos,'partidos')
```
